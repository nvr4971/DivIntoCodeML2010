Problem 1: Data acquisition

Read the iris dataset

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

Store the explanatory variables in variable X:

```python
X = pd.DataFrame(data = iris['data'], columns = iris['feature_names'])
print(X)
```

Store the objective variable in variable Y:

```python
Y = pd.DataFrame(data = iris['target'], columns = ['target'])
print(Y)
```

Problem 2: Combining data

Combined X and Y into variable df:

```python
df = pd.concat([X, Y], axis = 1)
print(df)
```

Problem 5: Extracting the required data

Extract sepal_width column in three different ways

```python
print(df['sepal width (cm)'])
print(df.loc[: , 'sepal width (cm)'])
print(df.iloc[: , 1])
```

Extract the 50th to 99th data

```python
print(df.iloc[49:99, :])
```

Extract the 50th to 99th data of the petal_length column

```python
print(df.loc[49:99, 'petal length (cm)'])
```

Extract data with a petal_width value of 0.2

```python
print(df.loc[df['petal width (cm)'] == 0.2])
```

Difference between loc and iloc:
- loc gets rows (or columns) with particular label (string) from the index
- iloc gets rows (or columns) at integers positions in the index

Problem 6: Creating a diagram

Make a pie chart of the number of samples per label (while showing percentages)

```python
import matplotlib.pyplot as plt

labels_count = df['target'].value_counts()
labels_count.plot.pie( 
    autopct = '%1.1f%%',
    startangle = 90,
    shadow = False,
    legend = True)
plt.show()
```

Select features one by one and visualize the distribution of data for each label using a violin plot

Difference between box plot and violin plot:

Violin plots are similar to box plots: a marker for the median of the data; a box or marker indicating the interquartile range; and possibly all sample points, if the number of samples is not too high. 

The difference is that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.
