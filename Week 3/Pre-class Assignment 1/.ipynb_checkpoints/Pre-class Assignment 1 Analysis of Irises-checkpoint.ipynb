Problem 1: Data acquisition

Read the iris dataset

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

Problem 2: Combining data

Combined X and Y into variable df:

```python
df = pd.concat([X, Y], axis = 1)
print(df)
```

Problem 4: Examining the iris dataset itself

The iris dataset (Fisher's Iris dataset) is a multivariate dataset introduced by the Birtish statistician, eugenicist and biologist Ronald Fisher in his 1936 paper "The use of multiple measurements in taxonomic problems" as an example of linear discriminant analysis.

The dataset consists of 50 samples from each of three species of Iris:
- Iris setosa (label 0)
- Iris versicolor (label 1)
- Iris virginica

Four features were measured from each sample (in cm):
- Length and width of the sepals
- Length and width of the 

Based on these four features, Fisher developed a linear discriminant model to distinguish the species from each other.

Difference between box plot and violin plot:

Violin plots are similar to box plots: a marker for the median of the data; a box or marker indicating the interquartile range; and possibly all sample points, if the number of samples is not too high. 

The difference is that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.
