Problem 1: Create random numbers

```python
import numpy as np

np.random.seed(0)

mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]

lst = np.random.multivariate_normal(mean, cov, 500)

print(lst)
```

Problem 2: Visualization by scatter plot

```python
import matplotlib.pyplot as plt

x = lst[:,0]
y = lst[:,1]

plt.scatter(x, y, marker='+')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

Problem 3: Visualization by histogram

```python
plot1 = plt.figure(1)
plt.hist(x)
plt.xlabel('x')
plt.ylabel('Frequency')
plt.xlim(-6, 3)

plot2 = plt.figure(2)
plt.hist(y)
plt.xlabel('y')
plt.ylabel('Frequency')
plt.xlim(-6, 3)

plt.show()
```

Problem 4: Addition of data

```python
np.random.seed(0)

mean1 = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]

lst1 = np.random.multivariate_normal(mean1, cov, 500)

x1 = lst1[:,0]
y1 = lst1[:,1]

mean2 = (0, -3)

lst2 = np.random.multivariate_normal(mean2, cov, 500)

x2 = lst2[:,0]
y2 = lst2[:,1]

plt.scatter(x1, y1, marker='+')
plt.scatter(x2, y2, marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend([0, 1])

plt.show()
```

Problem 5: Data combination

```python
lst = np.concatenate((lst1, lst2), axis = 0)

print(lst)
```

Problem 6: Labeling

```python
label1 = np.zeros((500, 1))
label2 = np.ones((500, 1))

label = np.concatenate((label1, label2), axis = 0)
lst = np.concatenate((lst, label), axis = 1)

print(lst)
```
